```python
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f,'green')
plt.title('График f(t)')
plt.xlabel('Значения f')
plt.ylabel('Значения t')
plt.axis([0.5,9.5, -2.5,2.5])
plt.show()
```

### Задание 3**

```python
x = np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x,y1)
ax1.set(xlim=(-5, 5))
ax1.set_title('График у1')

ax2.plot(x,y2)
ax2.set_title('График у2')

ax3.plot(x,y3)
ax3.set_title('График у3')

ax4.plot(x,y4)
ax4.set_title('График у4')
```

### Задание 4**

```python
import pandas as pd
data = pd.read_csv('creditcard.csv')
```

```python
plt.style.use('fivethirtyeight')
```

```python
data.head()
```

```python
count_class = data['Class'].value_counts()
count_class
```

```python
class_info = pd.Series(count_class)
class_info.plot(kind = 'bar')
plt.show()
```

```python
class_info.plot(kind = 'bar', logy = True)
plt.show()
```

```python
v1=data.set_index('Class')['V1'].filter(like='1', axis=0)
v1=v1.reset_index()
v1=v1.drop('Class', axis=1)
v1.head(), v1.count()
```

```python
v0=data.set_index('Class')['V1'].filter(like='0', axis=0)
v0=v0.reset_index()
v0=v0.drop('Class', axis=1)
v0.head(), v0.count()
```

```python
plt.hist(v0['V1'], bins=20, color='grey', edgecolor='black', density = True)
plt.hist(v1['V1'], bins=20, color='red', edgecolor='black', density = True)
plt.plot()
plt.xlabel('Class')
plt.legend(labels=['Class 0', 'Class 1'])
```
